# Loading dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Sachin Jorwal\OneDrive\Desktop\car sales.csv")
print(df)
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Drop irrelevant columns
columns_to_drop = ['Car_id', 'Customer Name', 'Dealer_No', 'Phone']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
```

# Univariate Analysis

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Price ($)'], bins=30, kde=True)
plt.title('Distribution of Car Prices')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Price ($)'])
plt.title('Boxplot of Car Prices')
plt.show()
```

## Countplot of Car Brands

```python
plt.figure(figsize=(12, 5))
sns.countplot(y=df['Company'], order=df['Company'].value_counts().index)
plt.title('Car Brand Popularity')
plt.show()
```

# Multivariate Analysis

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['Annual Income'], y=df['Price ($)'], hue=df['Gender'])
plt.title('Car Price vs. Annual Income')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='Transmission', y='Price ($)', data=df)
plt.title('Price Variation by Transmission Type')
plt.show()
```

```python
# Save cleaned dataset
df.to_csv("cleaned_car_sales.csv", index=False)
```
